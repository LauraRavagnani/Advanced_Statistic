```r
# A)
# simulation of the waiting time
lambda <- 1 / 30

random <- rexp(60, lambda)
#data <- data.frame(person = 1:60, waiting_time = random[order(random)])

# create the plot
#barplot(data$waiting_time, names = data$person, col = "pink", border = "magenta", space = 1, xlab = "person", ylab = "waiting time (min)", cex.lab = 1.5)

line <- par(lwd = 2)
hist(random, col = "pink", border = "magenta", xlab = "Waiting time (min)", cex.lab = 1.5)
```

```r
# B)
df <- data.frame(table(random))
df$random <- as.numeric(as.character(df$random))

less_12 <- with(df, sum(Freq[random < 12])) / 60

cat(paste("The probability, from the simulation, that a person will wait for less than 12 minutes is:", round(less_12 * 100, 2), "%", '\n'))
cat(paste("The probability, from the theory, that a person will wait for less than 12 minutes is:", round(pexp(12,lambda) * 100, 2), "%", '\n'))
```
