```r
library(ggplot2)

# a)
# write the pdf
dtriang <- function(x, a, c, b){
    pdf <- ifelse(x < c, 2*(x - a) / ((b - a) * (c - a)), 2*(b - x) / ((b - a) * (b - c)))
    return(pdf)
}

# plot the pdf
k <- 0:10
plot(k, dtriang(k, 0, 7, 10), type = "l", col = "purple", lwd = 2, cex.lab = 1.5, xlab = "x", ylab = "PDF")
grid(nx = NULL, ny = NULL, lty = 2, col = "gray", lwd = 0.5) 
```

```r
# b) function to generate random numbers
# cdf
ptriang <- function(x, a, c, b){
    val <- ifelse(x < c, ((x-a)^2)/((b-a)*(c-a)), (a*c - a*b - b*c + 2*b*x -(x^2))/((b-a)*(b-c)))
    return(val)
}

# function that samples from the pdf using the inverse method
rtriang <- function(n, a, c, b){
    u <- runif(n)
    val <- ifelse(u < (c - a)/(b - a), a + sqrt(u * (b - a) * (c - a)), b - sqrt(b^2 - a*b - b*c + a*c - u * (b - a) * (b - c)))
    return(val)
}
```

```r
# c)
# create the plot
df_line <- data.frame(k, dtriang(k, 0, 7, 10))
df_histo <- as.data.frame(rtriang(10^4, 0, 7, 10))

colors <- c("PDF" = "dark red", "Random numbers" = "blue")
ggplot(df_histo, aes(x = rtriang(10^4, 0, 7, 10))) +
    geom_histogram(aes(y = after_stat(density), col = "Random numbers",), binwidth = 0.5, fill = "light blue") +
    geom_line(df_line, mapping = aes(x = k, y = dtriang(k, 0, 7, 10), col = "PDF"), lwd = 1) +
    theme_bw() +
    labs(x = "x", y = "probability", color = "") +
    scale_color_manual(values = colors) +
    theme(legend.title = element_text(size = 17), legend.text = element_text(size = 15), legend.position = "top") +
    theme(axis.text = element_text(size = 13), axis.title = element_text(size = 15))
```
