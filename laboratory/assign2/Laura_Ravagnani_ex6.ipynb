```r
# annual return rate = gain/investment

investment <- 200 * 85
```

```r
# probability that after a year his net profit from the investment is at least 800e
prob <- 1 - pnorm(800, 0.1 * investment, 0.12 * investment)

cat(paste("The probability that after a year his net profit from the investment is at least 800€ is:", round(prob * 100, 2), "%", '\n'))
```

```r
# plot of the gain
x <- seq(-0.4 * investment, 0.6 * investment, 1)
y <- dnorm(x, 0.1 * investment, 0.12 * investment)

plot(x, y, type = "l", lwd = 2, xlab = "profit (eur)", ylab = "probability", cex.lab = 1.5)

x2 <- seq(800, 0.6 * investment, 1)
y2 <- dnorm(x2, 0.1 * investment, 0.12 * investment)
x2 = c(800, x2, 0.6 * investment)
y2 = c(0, y2, 0)
polygon(x2, y2, col = "orange", border = NA)
grid(nx = NULL, ny = NULL, lty = 2, col = "gray", lwd = 0.5)  
```
