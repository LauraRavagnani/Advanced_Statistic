```r
# model 
signal <- function(x, a, b, x0, w, t){
	t * (a * exp(-(x-x0)^2/(2*w^2)) + b)
}
```

Vary the sampling resolution $w$

```r
# model parameters
x0 <- 0 	    # signal peak
w <- c(0.1, 0.25, 1, 2, 3)			# signal width
a.real <- 2		# signal amplitude
b.real <- 1		# background amplitude
delta.t <- 5	#exposure time
```

```r
#set.seed(307)
par(mfrow = c(2, 3))

for(i in 1:length(w)){
	# generate observed data
	xdat <- seq(from = -7*w[i], to = 7*w[i], by = 0.5*w[i])
	signal.true <- signal(xdat, a.real, b.real, x0, w[i], delta.t)
	distrib_dat <- rpois(length(signal.true), signal.true)

	# calculate the log of the posterior on a grid {a_i, b_i}
	Nsamp <- 100
    a <- seq(0, 4, 4/Nsamp)
    b <- seq(0.5, 1.5, 1/Nsamp)
    
	# log posterior
	log.post <- function(d, x, a, b, x0, w, t){
		if(a < 0 || b < 0) {return(-Inf)}
		sum(dpois(d, lambda = signal(x, a, b, x0, w, t), log = TRUE))
	}

	logL <- matrix(0, nrow = length(a), ncol = length(b))
	for(j in 1:length(a)){
		for(k in 1:length(b)){
			logL[j,k] <- log.post(distrib_dat, xdat, a[j], b[k], x0, w[i], delta.t)
		}
	}

    # set maximum to zero and calculate the posterior
	logL <- logL - max(logL)
    post <- exp(logL)

	contour(a, b, post,
            nlevels = 5,
            labcex = 0.5,
            xlab = "a", ylab = "b",
            main = paste("w = ", w[i]))
	abline(v = a.real,
		col = "green3")
	abline(h = b.real,
		col = "green3")
}
```

Change the ratio $A/B$ used to simulate the data

```r
# model parameters
x0 <- 0 		# signal peak
w <- 1			# signal width
ratio <- c(0.5, 1, 1.5, 2, 2.5, 3)
a.real <- 1		# signal amplitude
b.real <- a.real * ratio		# background amplitude
delta.t <- 5	#exposure time
```

```r
par(mfrow = c(2, 3))
for(i in 1:length(ratio)){

	# generate observed data
	xdat <- seq(from = -7*w, to = 7*w, by = 0.5*w)
	signal.true <- signal(xdat, a.real, b.real[i], x0, w, delta.t)
	distrib_dat <- rpois(length(signal.true), signal.true)

	# calculate the log of the posterior on a grid {a_i, b_i}
	Nsamp <- 100
    a <- seq(0, 4, 4/Nsamp)
    b <- seq(0, 4, 4/Nsamp)

	# log posterior
	log.post <- function(d, x, a, b, x0, w, t){
		if(a < 0 || b < 0) {return(-Inf)}
		sum(dpois(d, lambda = signal(x, a, b, x0, w, t), log = TRUE))
	}

	logL <- matrix(0, nrow = length(a), ncol = length(b))
	for(j in 1:length(a)){
		for(k in 1:length(b)){
			logL[j,k] <- log.post(distrib_dat, xdat, a[j], b[k], x0, w, delta.t)
		}
	}

	# set maximum to zero and calculate the posterior
	logL <- logL - max(logL)
    post <- exp(logL)

	contour(a, b, post,
		nlevels = 5,
		labcex = 0.5,
		main = paste("ratio = ", ratio[i]),
        xlab = "a", ylab = "b")
	abline(v = a.real,
		col = "blue")
	abline(h = b.real[i],
		col = "blue")
}
```
