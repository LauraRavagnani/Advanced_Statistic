# Poisson process

```r
lambda <- 5 # baby/day
x <- 0:10
```

```r
# at least two babies in the next six hours
A <- dpois(x, lambda/4)
barplot(dpois(x, lambda/4), names = x)
```

```r
a <- 1 - sum(A[1:2])
cat(paste("the probability that at least two babies are born in the next six hours is", a))
```

```r
# no babies in the next two days
B <- dpois(x, 10)
barplot(B, names = x)
```

```r
cat(paste("no babies in the next two days:", B[1]))
```

# Exponential

```r
lambda <- 1/3   # earthquake/month
x <- 0:10
```

```r
prob <- dexp(x, lambda)
plot(x, prob, type = 'l')
```

```r
cat(paste("the probability that the next earthquake happens after three but before seven month is ", pexp(7, lambda) - pexp(3, lambda)))
```

# Electronic component

```r
lambda <- 1/10
x <- 0:15
prob <- dexp(x, lambda)
plot(x, prob, type = 'l')
```

```r
cat(paste("probability that lasts other 5 years is: ", 1-pexp(5, lambda)))
```
