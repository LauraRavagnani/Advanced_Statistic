# Birthday paradox

```r
n <- 365            # number of days
n.people <- 50      # number of people
r <- 1:n.people
```

```r
prob <- rep(0, n.people)
for(i in 2:n.people){
    n.tests <- 10^5
    common.birth <- 0
    for(j in 1:n.tests){
        # sample i random birthdays with repetition
        bdays <- sample(1:365, i, replace = TRUE)
        if(length(bdays) > length(unique(bdays))){
            common.birth <- common.birth + 1
        }
    }
    prob[i] <- common.birth / n.tests
}
```

```r
plot(r, prob, type = 'l')
```

# Poisson distribution

```r
lambda <- 1 #call/day
x <- 0:10
distrib <- dpois(x, lambda)

barplot(distrib, names = x)
```

```r
cat(paste("probability to get tomorrow 2 wrong calls is", distrib[3]))
```

```r
cat(paste("probability to get tomorrow at least 1 wrong call is", 1 - distrib[1]))
```

# Pascal

```r
ann_prob <- 0.58
maggie_prob <- 0.42
games <- 7
win <- 5
```

```r
# complete in 7 games means that ann won 4 games of 6 and then wins or maggie won 4 games of 6 and then wins
ann_win <- dbinom(win-1, games-1, ann_prob) * ann_prob
maggie_win <- dbinom(win-1, games-1, maggie_prob) * maggie_prob

ann_win + maggie_win
```

A: series ends in 7 games 


B: ann wins

$$ P(B|A) = \frac{P(AB)}{P(A)} $$

```r
ann_win/(ann_win+maggie_win)
```
